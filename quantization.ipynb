```python
from ultralytics import YOLO

model = YOLO("runs/ultralytics/finetune/yolov8n_lska_bifpn_EIoU_aug_all_10_640/triple_split/train/weights/best.pt", task='detect')  
results = model.predict("datasets/lard/detection/test_real_edge/images", conf=0.6, save=False, verbose=True)

infer_time = []
for result in results:
    infer_time.append(sum(list(result.speed.values())))
average_time = sum(infer_time)/len(infer_time)
print(f'平均推理时间: {average_time:.2f} ms')
```
